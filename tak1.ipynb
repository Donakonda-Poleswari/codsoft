```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('tested.csv')
```

```python
print(df)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
del df['Cabin']
```

```python
df
```

```python

df['Fare']=df['Fare'].fillna(df['Fare'].mean())
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
print(df)
```

```python
df.duplicated().sum()
```

```python
df['Embarked']=df['Embarked'].replace(['S','C','Q'],[0,1,2])
```

```python
df
```

```python
df=df.drop("Name",axis=1)
```

```python
df=df.drop("Ticket",axis=1)
```

```python
df=df.drop("PassengerId",axis=1)
```

```python
df['Sex']=df['Sex'].replace(['female','male'],[1,0])
```

```python
df['Age']=df['Age'].astype(int)
df['Fare']=df['Fare'].astype(int)
df
```

```python
df
```

```python
x=df.drop("Survived",axis=1)
x
```

```python
y=df.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=70)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

```python
y_predict=knn.predict(x_test)
```

```python
print(y_predict)
```

```python
accuracy=accuracy_score(y_predict,y_test)
print(accuracy)
```

```python
from sklearn.svm import SVC
support=SVC(kernel='poly',random_state=0);
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=70)
```

```python
support.fit(x_train,y_train)
```

```python
y_predict=support.predict(x_test)
print(y_predict)
```

```python
accuracy=accuracy_score(y_predict,y_test)
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
reg= RandomForestClassifier(n_estimators=500)
reg.fit(x_train,y_train)
y_predict=reg.predict(x_test)
#print(y_predict)
accuracy=accuracy_score(y_predict,y_test)
print(accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
y_predict=log.predict(x_test)
#print(y_predict)
accuracy=accuracy_score(y_predict,y_test)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
lis=[]
for i in range(1,100):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_predict=knn.predict(x_test)
    accuracy=accuracy_score(y_predict,y_test)
    lis.append(accuracy)
```

```python
age=df['Age']
```

```python
bins = [0, 30, 40,50,60,70]

categories = pd.cut(age, bins)
```

```python
counts=df.groupby([categories,df['Survived']]).size().unstack(fill_value=0)
```

```python
counts.plot(kind='bar',stacked=False,width=0.8)
plt.title('Age vs Outcome')
plt.xlabel('Age(in range)')
plt.ylabel('Survived(count)')
```
